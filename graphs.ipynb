```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

import glob
```

```python
df = pd.read_csv('csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv', \
                 parse_dates=True)
```

```python
df = df.groupby('Country/Region').sum()
```

```python
df = df.T
```

```python
df = df.drop('Lat')
df = df.drop('Long')
```

```python
df.columns
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df.index.name = 'Date'
```

```python
today = df.iloc[-1]
```

```python
print(today[(today >= 100) & (today < 1000)].index)
print(today[(today >= 1000) & (today < 10000)].index)
print(today[(today >= 10000)].index)
```

```python
usa = df['US']
usa100 = usa[usa > 100]
usa100 = usa100.reset_index()
usa100 = usa100['US']
usa100
```

```python
italy = df['Italy']
italy100 = italy[italy > 100]
italy100 = italy100.reset_index()
italy100 = italy100['Italy']
```

```python
france = df['France']
france100 = france[france > 100]
france100 = france100.reset_index()
france100 = france100['France']

germany = df['Germany']
germany100 = germany[germany > 100]
germany100 = germany100.reset_index()
germany100 = germany100['Germany']

spain = df['Spain']
spain100 = spain[spain > 100]
spain100 = spain100.reset_index()
spain100 = spain100['Spain']

swiss = df['Switzerland']
swiss100 = swiss[swiss > 100]
swiss100 = swiss100.reset_index()
swiss100 = swiss100['Switzerland']

skorea = df['Korea, South']
skorea100 = skorea[skorea > 100]
skorea100 = skorea100.reset_index()
skorea100 = skorea100['Korea, South']
```

```python
fig, axes = plt.subplots(1, sharex="all", gridspec_kw={'height_ratios': [1]}, figsize=(8, 5))

usa100.plot(logy=True)
italy100.plot(logy=True)
france100.plot(logy=True)
germany100.plot(logy=True)
spain100.plot(logy=True)
swiss100.plot(logy=True)
skorea100.plot(logy=True)

axes.legend()
axes.set_title("# of Confirmed COVID-19 cases in Western Countries (Log Scale)")
axes.set_xlabel("Days after passing 100 confirmed cases")
```

```python
fig, axes = plt.subplots(1, sharex="all", gridspec_kw={'height_ratios': [1]}, figsize=(8, 5))

usa100.plot()
italy100.plot()
france100.plot()
germany100.plot()
spain100.plot()
swiss100.plot()
skorea100.plot()

axes.legend()
axes.set_title("# of Confirmed COVID-19 cases in Western Countries (Polynomial Scale)")
axes.set_xlabel("Days after passing 100 confirmed cases")
```
